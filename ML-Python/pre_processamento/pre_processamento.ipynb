```python

import pandas as pd

#Definção dos títulos das colunas
headers = ['ESCT', 'NDEP', 'RENDA', 'TIPOR', 'VBEM', 'NPARC',
           'VPARC', 'TEL', 'IDADE', 'RESMS', 'ENTRADA', 'CLASSE']


#Carga do conjunto de treino
arquivo = 'https://raw.githubusercontent.com/MLRG-CEFET-RJ/ml-class/master/ml-t3/datasets/credtrain.txt'
data_train = pd.read_csv(arquivo, sep='\t', header=None, names=headers)

#Carga do conjunto de teste
arquivo = 'https://raw.githubusercontent.com/MLRG-CEFET-RJ/ml-class/master/ml-t3/datasets/credtest.txt'
data_test = pd.read_csv(arquivo, sep='\t', header=None, names=headers)



```

```python
import numpy as np

#Transoformação dos atributos e da classe alvo em matrizes
X_train_ = np.array(data_train.iloc[:, 0:11])
y_train_ = np.array(data_train['CLASSE'])


#Transformação dos atributos e da classe alvo em matrizes
X_test = np.array(data_test.iloc[:, 0:11])
y_test = np.array(data_test['CLASSE'])
```

```python
from sklearn.model_selection import train_test_split

#Separação de treino e validação
X_train, X_val, y_train, y_val = train_test_split(X_train_, y_train_,
                                                  train_size=0.8,
                                                  random_state=31)
```

```python
data_train.describe()
```

## Normalização dos dados

Como pode ser vistos nas saídas do final da seção as características estão em ordens de grandeza diferentes. Para muitos algoritmos este é um problema que pode levar a modelos inadequados. Para resolver este problema realizaremos a normalização dos dados.




```python
#Criação do transformador para noramlização
from sklearn.preprocessing import StandardScaler
norm = StandardScaler()

X_train_nor = norm.fit_transform(X_train)
X_test_nor = norm.fit_transform(X_test)
X_val_nor = norm.fit_transform(X_val)
```

```python
#Treinando o modelo com os dados normalizados
from sklearn.tree import DecisionTreeClassifier
modelo =  DecisionTreeClassifier(max_depth=3,random_state=31)
modelo.fit(X_train_nor, y_train)
```

```python
class_predict=modelo.predict(X_val_nor)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_val, class_predict))
```

## codificacao


codificacao esse processo consiste em condificar os atributos que sao categoricos nominais para um formator que seja mais adequado para o processo de treinamento e existem varias formas

label encoding-  que consistem em dar ou atribuir uma forma codifica de identificar uma determinada categoria ex:solteiro = 1 , casado = 2

count/frequecy encoding - consiste em codificar cada categoria consoante a sua frequencia ex: se o solteiro a parece 5x  no consujunto entao  todos campos com essa categoria sera passado para 5 e esse processo sera feito para as outras categorias, esse tipo de codificao e otimo quando se tem um numero elevado de categorias e tem uma limitacao quanto tem-se categorias com o mesmo numero de frequencia pois assim e perdido o sentido dos dados


one hat ecoding-consiste em criar mais colunas (o numero de categorias possiveis) que sao camadas de "variaveis dummies" em que  cada coluna representara aquela categoria expecifica em que nessas colunas teremos valor que oscilam entre 0 e 1 e caso aquela instancia  seja da quela categoria o seu valor sera 1 e o contrario 0
esse tipo de encodificao aumenta drasticamente a dimensao por isso aque se levar em consideram quando se tem um numero elevado de categorias

aqui sera usado o one hat encoding

## Aplicando as tecnicas de pre-processamento(codificao e normalizao) e treinamento

```python
# primeiro vou codificar tanto o conjunto de treino,validacao e o de teste

data_train_dumb= pd.get_dummies(data=data_train,
               prefix='EC',
               columns=['ESCT'],
               drop_first=False).astype(int)

data_test_dumb =  pd.get_dummies(data=data_test,
                                     prefix='EC',
                                      columns=['ESCT'],
                                       drop_first=False).astype(int)



```

```python
#vsu
data_train_dumb
```

```python
# separando os rotulos dos atibutos
X_train_ = np.array(data_train_dumb.drop(['CLASSE'], axis=1))
y_train_ = np.array(data_train_dumb['CLASSE'])

X_test = np.array(data_test_dumb.drop(['CLASSE'], axis=1))
y_test = np.array(data_test_dumb['CLASSE'])

```

```python
#separar os dados antes de normalizar para evitar o Data Leakage ou vazamento de dados pois para normalizacao dos conjuntos de teste e validacao usa-se transform
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train_, y_train_,
                                                  train_size=0.8,
                                                  random_state=31)

```

```python
#normalizando
from sklearn.preprocessing import StandardScaler
norm = StandardScaler()
#treino
train = norm.fit_transform(X_train)
#validacao
validation= norm.transform(X_val)
#teste
test = norm.transform(X_test)


```

```python
#treinando
from sklearn.tree import DecisionTreeClassifier
modelo =  DecisionTreeClassifier(max_depth=3,random_state=31)
modelo.fit(train,y_train)
```

```python
#validando
class_predict=modelo.predict(validation)
```

```python
#vendo o desenpecnhofrom sklearn.metrics import classification_report
print(classification_report(y_val, class_predict))
```

```python
#usando o modelo sobre o conjunto de teste
class_predict_test=modelo.predict(test)
```

```python
#vendo o desempenho
print(classification_report(y_test, class_predict_test))
```
