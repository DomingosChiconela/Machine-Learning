usando  tensor  escalar ou seja de  dimensao  0

```python
a= tf.constant(2,name="n1")
b= tf.constant(4,name="n2")
c= tf.constant(6,name="n3")
a,b,c,
```

```python
i1 =tf.multiply(a,b)
i2 = tf.subtract(a,c)
```

```python
saida =  tf.add(i1,i2)
saida
```

Usando  tensor de matricial

```python
n1 = tf.constant([[1.0,2.0],[3.0,4.0]])
n2 = tf.constant([[8.0,6.0],[4.0,2.0]])
n3 =  tf.constant([[10.0,12.0],[9.0,13.0]])

intern1 = tf.matmul(n1,n2)
intern2 = tf.subtract(n1,n3)
saida = tf.add(intern1,intern2)
saida
```
