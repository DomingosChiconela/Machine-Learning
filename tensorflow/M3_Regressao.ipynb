# Regressao
Os modelos de regressao  buscam  descrever as  relacoes  que existem  entre  variaveis

```python
%tensorflow_version 2.x
```

```python
import  tensorflow as  tf
import numpy  as np
import matplotlib.pyplot as plt
```

```python
peso = [[3.0]]
bias = [0.5]
nivel_ruido = 0.5
tamanho_lote  =  100

feature =  tf.random.normal(shape= [tamanho_lote,tf.shape(peso)[0]])
target  =  tf.matmul(feature,peso) + bias + nivel_ruido * tf.random.normal(shape=tf.shape(feature))
```

```python
plt.plot(feature,target,'ro')
plt.xlabel('Sensacao  termica')
plt.ylabel('Consumo de energia')
plt.title('Dados para Regressao')
plt.plot(np.unique(feature), np.poly1d(np.polyfit(tf.squeeze(feature),tf.squeeze(target),1))(np.unique(feature)))
plt.show()
```
